```python
def cost_function(X,y,B):
    return (np.sum((np.dot(X,B) - y) ** 2)) / (2 * n)
```

```python
cost_function(X,y,B)
```

```python
def gradient_descent(X,y,B,epochs,alpha):
    cost_history = [0] * epochs
    for i in range(epochs):
        y_pred = X.dot(B)
        loss = y_pred - y
        #gradient descent - minimising the loss
        gradient = X.T.dot(loss) / n
        B = B - alpha * gradient
        print("B =",B)
        cost_history[i] = cost_function(X,y,B)
#         print(i)
    return cost_history, B
```

```python
cost,b = gradient_descent(X,y.flatten(),B,epochs=5000,alpha=0.0001)
```
